```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Computer Science\Data Analytics\Diwali Sales\Diwali Sales Data.csv',encoding='unicode_escape')
df
# To avoid encoding error we use 'unicode_escape'
```

```python
df.drop(columns = ['Status','unnamed1'],axis=1,inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df
```

```python
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.describe()
```

# Exploratory Data Analysis

```python
df.columns
```

```python
gen = sns.countplot(x='Gender',data = df)

for bars in gen.containers:
    gen.bar_label(bars)
```

```python
amt = df.groupby('Gender')[['Amount']].sum()
amt
```

```python
sns.barplot(x = amt.index,y = 'Amount',data = amt)
```

Most of the Buyers are Female and purchasing power of women is greater than Men

```python
shopping_agegroup = sns.countplot(x = 'Age Group',data = df,hue = 'Gender')

for bars in shopping_agegroup.containers:
    shopping_agegroup.bar_label(bars)
```

```python
sales_age = df.groupby('Age Group')[['Amount']].sum().sort_values(by = 'Amount',ascending = False)
sales_age
```

```python
sns.barplot(x = sales_age.index,y = 'Amount',data = sales_age)
```

From above graphs we can say that most of the buyers are of age group 26-35 in which mostly are Females

```python
states_ord = df.groupby('State')[['Orders']].sum().sort_values(by = 'Orders',ascending=False)
states_ord
```

```python
plt.figure(figsize= (20,10))
plt.xticks(rotation = 30,ha= 'right')
sns.barplot(x = states_ord.index,y = 'Orders',data = states_ord)
```

Most of the order came from State Uttar Pradesh,Maharastra and Karnataka respectively

```python
states_amt = df.groupby('State')[['Amount']].sum().sort_values(by = 'Amount',ascending = False)
states_amt
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation = 30,ha = 'right')
sns.barplot(x = states_amt.index,y = 'Amount',data = states_amt)
```

Most of the Money spent on shopping is from state Uttar Pradesh

```python
df.columns
```

```python
marital_amt = df.groupby(['Marital_Status','Gender'],as_index=False)[['Amount']].sum().sort_values(by = 'Amount',ascending = False)
marital_amt
```

```python
sns.barplot(x = 'Marital_Status',y = 'Amount',data = marital_amt,hue = 'Gender')
```

From the above Graph we can say that most of the buyers are Married Women

```python
plt.xticks(rotation = 30,ha = 'right')
occ_amt = df.groupby('Occupation')[['Amount']].sum().sort_values(by = 'Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
occ_amt_val = sns.barplot(x = occ_amt.index,y = 'Amount',data = occ_amt)
```

Most of the Buyers are working in IT sector,Healtcare,Aviation respectively

```python
plt.xticks(rotation = 30,ha = 'right')
prod_cat_ord = df.groupby('Product_Category')[['Orders']].sum().sort_values(by='Orders',ascending = False)
sns.barplot(x = prod_cat_ord.index,y ='Orders',data = prod_cat_ord)
```

```python
plt.xticks(rotation = 30,ha = 'right')
prod_cat_amt = df.groupby('Product_Category')[['Amount']].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x = prod_cat_amt.index,y ='Amount',data = prod_cat_amt)
```

Product Category which is most ordered is Clothing and product category on which most of the money spent is Food

```python
top_10_selling_prod = df.groupby('Product_ID')[['Orders']].sum().sort_values(by = 'Orders',ascending = False).head(10)
top_10_selling_prod
```

```python
sns.barplot(x = top_10_selling_prod.index,y= 'Orders',data= top_10_selling_prod)
```

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
